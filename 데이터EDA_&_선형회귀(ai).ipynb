#산점도 예시
Training compute (FLOP) × Training dataset size (datapoints)

의미: 학습량과 학습 데이터 포인트 수 → 모델 학습 규모와 데이터 규모 관계

Training compute (FLOP) × Training time (hours)

의미: 학습량과 학습 시간 → 계산량 대비 학습 시간 효율성

Training compute (FLOP) × Training compute cost (2023 USD)

의미: 학습량과 비용 → 계산량 증가에 따른 비용 증가

Training dataset size (datapoints) × Citations

의미: 데이터셋 크기와 논문 영향력 → 더 큰 데이터로 학습한 모델이 더 인용되는지 확인

Training time (hours) × Training compute (FLOP)

의미: 학습 시간과 계산량 → 시간 대비 계산량 관계, 효율성 분석

Training power draw (W) × Training compute (FLOP)

의미: 계산량과 소비 전력 → 학습 효율성과 전력 사용 분석

Finetune compute (FLOP) × Batch size

의미: 파인튜닝 계산량과 배치 크기 → 배치 크기가 파인튜닝 효율에 미치는 영향

Finetune compute (FLOP) × Training time (hours)

의미: 파인튜닝 계산량과 소요 시간 → 시간 대비 파인튜닝 효율

Batch size × Training time (hours)

의미: 배치 크기와 학습 시간 → 배치 크기에 따른 학습 시간 변화

Citations × Training compute (FLOP)

의미: 모델 학습량과 논문 영향력 → 더 큰 모델이 연구 커뮤니티에서 더 많이 인용되는지 확인

Training compute cost (2023 USD) × Training time (hours)

의미: 학습 비용과 시간 → 시간 증가가 비용에 미치는 영향

````python
#계산량과 학습에 걸린 시간 산점도를 그려보기
import matplotlib.pyplot as plt
plt.scatter(df['Training compute (FLOP)'],df['Training compute cost (2023 USD)'])
plt.xlim((0, 1e25))
plt.savefig("산점도.pdf")
plt.show()
````
